```python
import base64
import time
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
import os

# Set up headless Chrome
chrome_options = Options()
chrome_options.add_argument('--headless')
chrome_options.add_argument('--disable-gpu')
driver = webdriver.Chrome(options=chrome_options)

# Define the range of chapters
start_chapter = 113
end_chapter = 122
base_url = "https://www.natomanga.com/manga/vinland-saga/chapter-{}"

for chapter in range(start_chapter, end_chapter + 1):
    url = base_url.format(chapter)
    driver.get(url)
    
    # Initial load and scroll
    time.sleep(5)
    last_height = driver.execute_script("return document.body.scrollHeight")
    while True:
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        time.sleep(2)
        new_height = driver.execute_script("return document.body.scrollHeight")
        if new_height == last_height:
            break
        last_height = new_height
    
    # Final dimensions check
    time.sleep(2)
    total_height = driver.execute_script("return document.documentElement.scrollHeight")
    total_width = driver.execute_script("return document.documentElement.scrollWidth")
    
    # Adjust window size
    driver.set_window_size(total_width, total_height)
    time.sleep(2)
    
    # Recheck dimensions after resize
    total_height = driver.execute_script("return document.documentElement.scrollHeight")
    total_width = driver.execute_script("return document.documentElement.scrollWidth")
    
    # Calculate PDF dimensions in inches (96 DPI)
    paper_width = total_width / 96
    paper_height = total_height / 96
    
    # PDF generation parameters
    pdf_params = {
        "printBackground": True,
        "landscape": False,
        "displayHeaderFooter": False,
        "marginTop": 0,
        "marginBottom": 0,
        "marginLeft": 0,
        "marginRight": 0,
        "paperWidth": round(paper_width, 2),
        "paperHeight": round(paper_height, 2),
        "scale": 1,
    }
    
    # Generate and save PDF
    pdf = driver.execute_cdp_cmd("Page.printToPDF", pdf_params)
    if not os.path.exists("vinland saga"):
        os.makedirs("vinland saga")
    output_file = f"vinland saga/chap_{chapter}.pdf"
    with open(output_file, 'wb') as f:
        f.write(base64.b64decode(pdf['data']))
    
    print(f"Downloaded chapter {chapter} as single-page PDF")

driver.quit()
```
